```python
# separating numerical and non-numerical data

df_text = df2[['Parameter', 'Colour', ' Odour ']]
df_text
```

```python
df_num = df2.drop(['Colour', ' Odour '], axis = 1)
df_num
```

```python
df_num.dtypes
```

```python
df_temp = df2[['Parameter', 'Turbidity(NTU)']]
```

```python
df_temp = df_temp.drop(df_temp.index[[0,1]])
df_temp
```

```python
sns.set_style("ticks")
```

```python
# Turbidity

plt.subplots(figsize = (15,10))
sns.set_context("notebook")
sns.barplot(y = df_temp['Turbidity(NTU)'], x = df_temp['Parameter'])
plt.xlabel('Place')
plt.ylabel('Turbidity(NTU)')
plt.xticks(rotation = 45)
plt.axhline(y=5, color='b', linestyle='-')
plt.axhline(y=10, color = 'r', linestyle='-')
```
